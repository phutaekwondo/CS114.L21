##1.1. Giới thiệu (đặt vấn đề)

Ở Việt Nam hiện nay, trong các loại phương tiện di chuyển thì xe máy là phổ biến nhất. Do đó việc quản lý, trông giữ xe máy trở nên rất quan trọng, từ các cơ quan hành chính, trường học, siêu thị,... đến quán cafe, quán ăn, nhà nghỉ đều có khu vực giữ xe. 

Thông thường chổ gữi xe sẽ quản lý bằng cách gán cho mỗi xe của khách một thẻ ( thẻ này có thể là thẻ điện tử hoặc thẻ viết tay của bảo vệ ), khách sẽ giữ thẻ này đến khi muốn lấy xe thì đưa thẻ cho người trông xe. Việc nhận biết thẻ có đúng với xe hay không sẽ được người giữ xe thực hiện bằng cách đọc biển số xe ghi trên thẻ, và biển số xe thực tế của xe. Nếu giống nhau thì cho khách lấy xe về.
